1定義函數a(x,y)
2呼叫函數a(x,y)

```python
def a(x,y):
  result = x + y*2
  return result


b = a(1,2)
c = a(2,3)
print(b)
print(c)
```

while迴圈
定義函數a(x),並呼叫兩次函數

```python
def a(x):
  result = 0
  while result < 10:
    result = result + x
    if result==5:
      return result
  return result


b = a(1)
c = a(2)
print
```

```python
def a(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n + a(n-1)

print(a(3))
```
